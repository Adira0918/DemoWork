### Task 8

####  Mobile & Travel Insurance premium are discounted by 10% 
#### Health and Property Insurance premium are increased by 7%
#### Life and Motor Insurance premium are marginally increased by 2%
#### What will be overall change in % of the Premium Amount Collected for all these Customer?

```python
all_data.INSURANCE_TYPE.unique()
```

```python
all_data['Changed_Amount'] = 0 
all_data.loc[
       (all_data['INSURANCE_TYPE'] == 'Health'),
       'Changed_Amount'] = all_data['PREMIUM_AMOUNT']+((all_data['PREMIUM_AMOUNT']*7)/100)
all_data.loc[
       (all_data['INSURANCE_TYPE'] == 'Property'),
       'Changed_Amount'] = all_data['PREMIUM_AMOUNT']+((all_data['PREMIUM_AMOUNT']*7)/100)
all_data.loc[
       (all_data['INSURANCE_TYPE'] == 'Mobile'),
       'Changed_Amount'] = all_data['PREMIUM_AMOUNT']-((all_data['PREMIUM_AMOUNT']*10)/100)
all_data.loc[
       (all_data['INSURANCE_TYPE'] == 'Travel'),
       'Changed_Amount'] = all_data['PREMIUM_AMOUNT']-((all_data['PREMIUM_AMOUNT']*10)/100)
all_data.loc[
       (all_data['INSURANCE_TYPE'] == 'Life'),
       'Changed_Amount'] = all_data['PREMIUM_AMOUNT']+((all_data['PREMIUM_AMOUNT']*2)/100)
all_data.loc[
       (all_data['INSURANCE_TYPE'] == 'Motor'),
       'Changed_Amount'] = all_data['PREMIUM_AMOUNT']+((all_data['PREMIUM_AMOUNT']*2)/100)
```

```python
all_data.sort_values(by='TXN_DATE_TIME',ascending=True).reset_index()
```

```python
all_data['sales_pct_changes'] = all_data['Changed_Amount'].pct_change()
```

```python
#Overall Percent change
all_data
```
